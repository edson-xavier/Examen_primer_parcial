```
print("PROGRAMA PARA OBTENER RAICEZ MEDIANTE LA FORMULA GENERAL")
print("ECUACIONES DEL TIPO: ax^2+bx+c")
import math

disc = -1
while disc < 0:
  a = int (input("INGRESE EL VALOR DE ax^2:"))
  b = int (input("INGRESE EL VALOR DE bx:"))
  c = int (input("INGRESE EL VALOR DE c:"))

  disc = (b**2)-(4 * a * c)

  if disc < 0:
    print("ERROR, EL DISCRIMINANTE TIENE QUE SER POSITIVO")

  res1= (-1) * b

res2= math.sqrt(disc)
res3= (res1 + res2)/(2 * a)
res4= (res1 - res2)/(2 * a)

print("los valores de x son:")
print("X1 = ", res3)
print("X2 = ", res4)

```
